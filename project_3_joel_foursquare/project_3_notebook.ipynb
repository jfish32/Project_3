```python
import requests
import pandas as pd
from sqlalchemy import create_engine
import pprint
```

```python
url = 'https://api.foursquare.com/v3/places/search'
```

```python
headers = {
    'Authorization': 'fsq32FCDhz9f+pnuttA0LvVLtxSGHrMROavrcMWxKP+u/4M=',
    'Accept': 'application/json'
}
```

```python
params = {
    'near': 'Chicago',    # Location to search near
    'query': 'restaurant',      # Search query
    'limit': 50                 # Number of results to return

}
```

```python
response = requests.get(url, params=params, headers=headers)
print(response)
```

```python
data =response.json()

data['results'][0]['geocodes']['main']
```

```python
restaurants = []
for venue in data['results']:
    name = venue['name']
    location = venue['location']['formatted_address']
    description = ', '.join(category['name'] for category in venue['categories'])
    latitude = venue['geocodes']['main']['latitude']
    longitude = venue['geocodes']['main']['longitude']
    restaurants.append({'name': name, 'location': location,'latitude':latitude,'longitude':longitude, 'description': description,'city':'CHICAGO','state':'IL'})
```

```python
df_chicago = pd.DataFrame(restaurants)
df_chicago
```
